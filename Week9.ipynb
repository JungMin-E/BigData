```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

from sklearn.datasets import load_breast_cancer
b_cancer = load_breast_cancer()
b_cancer
```

```python
b_cancer.keys()
```

```python
df = pd.DataFrame(b_cancer.data, columns = b_cancer.feature_names)
df['diagnosis'] = b_cancer.target
df.head()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
b_cancer_scaled = scaler.fit_transform(b_cancer.data)
b_cancer_scaled
```

```python
pd.DataFrame(b_cancer_scaled).describe()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

Y = b_cancer.target
X = b_cancer_scaled

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
X_train.shape[0], X_test.shape[0]
```

```python
lr_b_cancer = LogisticRegression()
lr_b_cancer.fit(X_train, Y_train)
Y_predict = lr_b_cancer.predict(X_test)
Y_predict
```

```python
from sklearn.metrics import confusion_matrix

#혼동 행렬(CrossMatrics)만들기
cm = confusion_matrix(Y_test, Y_predict)
cm
```

```python
TN, FP, FN, TP = cm[0, 0], cm[0, 1], cm[1, 0], cm[1, 1]

accuracy = (TN + TP) / (TN + TP + FN + FP)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1 = 2 * ((precision*recall) / (precision+recall))

print(accuracy, precision, recall, f1)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

accuracy = accuracy_score(Y_test, Y_predict)
precision = precision_score(Y_test, Y_predict)
recall = recall_score(Y_test, Y_predict)
f1 = f1_score(Y_test, Y_predict)
roc_auc = roc_auc_score(Y_test, Y_predict)

print(f'정확도: {accuracy:.3f}, 정밀도: {precision:.3f}, 재현율: {recall:.3f}, F1: {f1:.3f}')
print(f'ROC_AUC: {roc_auc:.3f}')
```
