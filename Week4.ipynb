```python
#Week4 2nd Class
import pandas as pd
df = pd.read_csv("seoul_apartment.csv")
df.head()
```

```python
df.info()
```

```python
pd.set_option("display.float_format", "{:.0f}".format)
df.describe()
```

```python
df.sort_values(by="층")
```

```python
df.groupby("시군구")["거래금액(만원)"].mean()
df["평균거래금액"] = df.groupby("시군구")["거래금액(만원)"].transform("mean")
df
```

```python
df.groupby("시군구").filter(lambda x: x[ "거래금액(만원)"].mean() >= 300000)
```

```python
df.groupby("시군구")["거래금액(만원)"].mean().reset_index().sort_values(by="거래금액(만원)", ascending=False).head(5)
```

```python
#reset_index 인덱스를 일반적인 데이터 열(column)으로 바꾸고 기존 인덱스는 기본정수 인덱스(0, 1, 2)로 교체
df.groupby("건축년도")["거래금액(만원)"].mean().reset_index() 
```
