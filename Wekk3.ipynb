```python
import pandas as pd
pd.__version__
```

```python
df = pd.read_csv("data/example.csv", index_col=0)
df
```

```python
#결측치 제거(열을 제거, 행을 제거)
#dropna(): 결측치가 포함된 행 전체를 제거
#DataFRame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False) inplace = True로 설정하면 결측치를 제외한 값들을 업데이트해서 반환
df.dropna(axis=1) #1열 반환
```

```python
#df.fillna(value=None, axis=None, inplace=False, limit=None)
#value는 결측치를 대체할 값, axis는 결측치를 대체할 방향을 지정(0:행, 1:열), limit 한번에 대체할 결측치의 개수를 지정

df['age'] = df['age'].fillna(df['age'].mean())
df
```

```python
#결측치 대체
#ffill(): 이전 값으로 결측치 채움
#bfill(): 이후 값으로 결측치 채움

df = df.ffill()
df #결측치의 이전 값인 Daejeon으로 값 설정.
```

```python
df.dtypes['age'] #해당 열의 dataType을 확인할 수 있는
```

```python
#astype(dtype, copy=True, error='raise') error = ignore이면 에러 무시 error = raise(기본값)이면  변활할 수 없는 값이 
df['age'].astype(int) #df의 age열의 type을 int로 변환
```

```python
#열 이름 변경 rename()
#DataFrame.rename(index=None, column=None, axis=None, copy=True, inplace=False, level=None, error='ignore')
```

```python
#데이터 정렬
df.sort_values(by='age', ascending=True, inplace=True)
df
```

```python
import pandas as pd #pandas를 pd라는 별칭으로 

df = pd.read_csv('data/student_admission_record_dirty.csv')
df
```

```python
df = df.sort_values(by='Age')
df
```

```python
#중복 데이터 처리 drop_duplicates(subset=None, keep='first', inplace=False)
df = df.drop_duplicates(subset='Age')
df
```

```python
#이상치 처리
df = df[df['Age'] > 0]
df
```

```python
#문자열 데이터 정제
#str.strip(): 문자열의 양쪽 끝에 있는 불필요한 공백 제거
#str.replace(): 문자열 내 특정 부분을 다른 문자열로 대체
#str.upper(): 문자열을 대문자로 변환
#str.lower(): 문자열을 소문자로 변환

df['Name'].str.lower()
```

```python
#concat()
#obj: 연결할 데이터프레임, 리스트 또는 dictoniary 형태로
#
import pandas as pd
df1 = pd.DataFrame({"id":[1,2,3], "name":["Alice", "Bob", "Charlie"]})
df2 = pd.DataFrame({"id":[2,3,4], "score":[85,90,95]})
print(df1, df2)
```
