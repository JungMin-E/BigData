```python
import pandas as pd
pd.set_option('display.max_rows', 15)
df = pd.read_csv('student_admission_record_dirty.csv') 
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum() #결측치 확인방법
```

```python
df[df['Age'].isnull()] #Name Column에서 결측치 있는 값들 다 찾아보기
df['Age'] = df['Age'].fillna(df['Age'].mean()) #Age결측치의 값들을 모두 Age의 평균값으로 대체하기
```

```python
print(df.columns)
```

```python
df[df['Name'].isnull()]
df['Name'] = df['Name'].ffill() #이름의 결측치 값들을 모두 이전 값들로 대체하기
df
```

```python
df[df['Gender'].isnull()]
df['Gender'] = df['Gender'].bfill() #Gender 결측치 값들 이후 데이터 값들로 대체하기
df
```

```python
df[df['Admission Test Score'].isnull()]
df['Admission Test Score'] = df['Admission Test Score'].fillna(df['Admission Test Score'].median()) #Admission Test Score 값들의 중앙값으로 대체하기
df
```

```python
df.isnull().sum()
```

```python
df[df['High School Percentage'].isnull()]
df['High School Percentage'] = df['High School Percentage'].fillna(df['High School Percentage'].median()) #High School Percentage 값들의 중앙값으로 대체하기
print(df)
```

```python
df.isnull().sum()
df['City'].mode()[0] #City에서 가장 많이 나온 도시는 "Quetta"임
df['City'] = df['City'].fillna(df['City'].mode()[0]) #City에서 가장 많이 나온 도시인 "Quetta"를 최빈값으로 하여 해당 값을 결측치로 대처함
```

```python
df.isnull().sum()
df[df['Admission Status'].isnull()]
df['Admission Status'] = df['Admission Status'].ffill() #Admission Status를 이전 값으로 대처하기 그 이유는 결측치들을 보았을 때
#해당 값이 66, 67로 연속적인 결측치가 존재하였지만 65번째 값으로 결측치를 대채하기
```
