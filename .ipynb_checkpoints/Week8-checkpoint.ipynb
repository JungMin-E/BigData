```python
import pandas as pd
df = pd.read_csv("8강데이터/auto-mpg.csv")
print(df.shape)
df.head()
```

```python
df.describe()
```

```python
df = df.drop(['car_name', 'horsepower', 'origin'], axis=1) #axis=1열 삭제
# df['mpg', 'cylinders', 'displacement', 'weight', 'acceleration', 'model_year']
df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
y = df['mpg']
X = df.drop(['mpg'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
lr = LinearRegression()
lr.fit(X_train.values, y_train)
```

```python
y_predict = lr.predict(X_test.values)
```

```python
y_predict
```

```python
import numpy as np
mse = mean_squared_error(y_test, y_predict)
rmse = np.sqrt(mse) 
r2 = r2_score(y_test, y_predict)

print(mse)
print(rmse)
print(r2)
```

```python
print("연비 예측하기")

cylinder = float(input("cylinder:  "))
displacement = float(input("displacement:  "))
weight = float(input("weight:  "))
acceleration = float(input("acceleration:  "))
model_year = float(input("model_year:  "))

mpg_predict = lr.predict([[cylinder, displacement, weight, acceleration, model_year]])
mpg_predict
```

```python
#실제 값에 대한 예측 값 그래프를 나타내고 얼마나 다른지 시각화로 비교하기
import matplotlib.pyplot as plt
import numpy as np

plt.scatter(y_test, y_predict, alpha=0.6)
min_val = y_test.min() #10
max_val = y_test.max() #46

plt.xlabel("Actual MPG")
plt.ylabel("Predicted MPG")
plt.title("Actual vs Predicted MPG")

plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--')
plt.show()
```

```python
#실습
import pandas as pd

df = pd.read_csv("대기오염데이터_종로구_20220401_20240215.csv", encoding='cp949')
df.shape
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

#df['so2'] 결측치 평균값으로 대체

df['so2'] = df['so2'].fillna(df['so2'].mean())
df.isnull().sum()
```

```python
#df['location']열 삭제
df = df.drop('location', axis=1)
df
```

```python
df['day'] = pd.to_datetime(df['day'])
            
df['year'] = df['day'].dt.year
df['month'] = df['day'].dt.month
df['day2'] = df['day'].dt.day
```

```python
#1. X: 원본 날짜 열은 반드시 제외!
features = ['so2', 'no2', 'co', 'o3', 'day2']
X = df[features]

#2. Y
y = df['pm10']

#3. 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
#Linear_model 활용해서 회귀 분석 모델 학습
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_predict = lr.predict(X_test.values)
y_predict
```

```python
#MSE, RMSE, R2_score 출력 및 결과 분석
import numpy as np

mse = mean_squared_error(y_test, y_predict)
rmse = np.sqrt(mse) 
r2 = r2_score(y_test, y_predict)

print(mse)
print(rmse)
print(r2)
```

```python
#seaborn의 regplot()으로 독립변수와 종속변수 간 관계 그래프 그리기

import matplotlib.pyplot as plt

plt.scatter(y_test, y_predict, alpha=0.6)
min_val = y_test.min() #10
max_val = y_test.max() #46

plt.plot([min_val, max_val], [min_val, max_val], color='b', linestyle='--')
plt.show()
```
