```python
import pandas as pd
import numpy as np
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from functools import reduce

words = []
for title in df['제목']:
    #1. 정제
    result = re.sub(r"[^a-zA-Z]+", " ", str(title))
    #2. 단어 토큰화
    result = word_tokenize(result.lower())
    #3. 불용어 제거
    result = [w for w in result if w not in stopWords]
    #4. 표제어 추출
    result = [lemma.lemmatize(w) for w in result]
    #5.데이터 최종정리
    words.append(result)


#2차원 리스트를 1차원 리스트로 변환
word_list = list(reduce(lambda x, y: x+y, words))
#word_list = list(np.array(words, dtype=object).reshape(-1,1))
#word_list
```

```python
#단어빈도 분석 collections package사용
from collections import Counter

count = Counter(word_list)

#most_common() collections package에 포함된 빈도수가 높은 상위 N개의 list로 반환
word_count = count.most_common(50)
word_count = dict(word_count)

print(word_count)
```

```python
import matplotlib.pyplot as plt

plt.bar(word_count.keys(), word_count.values())
plt.xticks(rotation=85)
#plt.bar(5,20)
plt.show()
```

```python
#시각화 2.
from wordcloud import STOPWORDS, WordCloud

stopwords = set(STOPWORDS)
wc = WordCloud(background_color = 'ivory',
               stopwords = stopwords,
               width = 800,
               height = 600)

#del 사용해서 딕셔너리 값안의 'data', 'big'값을 제거
del word_count['data'] 
del word_count['big']

cloud = wc.generate_from_frequencies(word_count)
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
