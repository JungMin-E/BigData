```python
import pandas as pd
import numpy as np
df = pd.read_csv('students_raw.csv') 
df
```

```python
df.info()
```

```python
#문제 1.
df_columns = len(df.columns) #df의 column의 수
df_rows = len(df) #df의 rows의 수
print(f"열: {df_columns}, 행: {df_rows}")
```

```python
#문제 2.
df = df.astype({'student_id': 'str', 
                'enrollment_year': 'str'})
#student_id와 enrollment_year열 str타입으로 변환
df.info()
```

```python
#.str.strip() 양쪽의 공백 제거 
#.str.capitalize() 첫글자 대문자
df['name'] = df['name'].str.strip().str.capitalize() 
df['city'] = df['city'].str.strip().str.capitalize()
df
```

```python
#문제 3.
#.str.strip()앞, 뒤 모든 공백을 제거
#.str.lower() 문자열을 소문자로 변환
df['email'] = df['email'].str.strip().str.lower()
df['major'] = df['major'].str.strip() 
#.str.contains('값', na=False) 해당 값이 포함되어 있지 않다면 False
#r'\.com' .은 아무 글자 하나를 의미하기에 r''은 이 문자열을 정규표현식으로 사용한다는
#np.where(질문, yes! 할일, 아니오! 할일)구조이다.
df['invalid_email'] = np.where(
    (df['email'].isnull()) | 
    (~df['email'].str.contains('@', na=False)) |
    (~df['email'].str.contains(r'\.com', na=False)), 1, 0
)
df
```

```python
#문제 4.
#rename(colmuns, rows={'원래이름': '바꿀이름'})
df = df.rename(columns={'height_cm': 'height', 'enrollment_year': 'enroll_year'})
df
```

```python
#문제 5.
#dropna(subset=['기준열']) 기준열에 결측치가 존재하는 행 전체를 제거
#drop_duplicates(['기준열']) 기준열에 존재하는 중복된 값을 해당 행을 제거
df['student_id'] = df['student_id'].replace('nan', np.nan) #dropna함수는 np.nan만 찾도록 되어있어
df = df.dropna(subset=['student_id'])
df = df.drop_duplicates(subset=['student_id'])
df = df.sort_values(by='gpa', ascending=False)
df               
```

```python
#문제 6.
df = df[df['gpa'].between(0, 4.5)] #gpa의 값 0~4.5사이의 값들만 남기기
df = df.dropna(subset=['gpa']) #gpa열에 결측치가 있는 행을 제거
df
```

```python
#문제 7.
#pd.merge(데이터프레임1, 데이터프레임2, on='공통되는 열의 이름', how='left', 'right'방식 선택)
major_df = pd.read_csv('major_lookup.csv')
major_df
df_merged = pd.merge(df, major_df, left_on='major', right_on='raw_major', how='left')
df = df_merged.rename(columns={'normalized_major': 'new_major'})
df
```
